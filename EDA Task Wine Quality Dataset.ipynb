```python
numeric_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
len(numeric_cols)
```

```python
my_figure = plt.figure(figsize=(12,9))
title = my_figure.suptitle("Numeric columns distribution")
for i in numeric_cols:
    plt.subplot(4,3,numeric_cols.index(i)+1)
    plt.hist(x=df[i])
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.tight_layout()
plt.show()
```

```python
my_figure = plt.figure(figsize=(16,12))
title = my_figure.suptitle("Numeric columns Outlier detection")
for i in numeric_cols:
    plt.subplot(4,3,numeric_cols.index(i)+1)
    plt.boxplot(x=df[i])
    plt.xlabel(i)
    plt.tight_layout()
plt.show()
```

```python
df_corr = df[numeric_cols].corr()
mask = np.triu(np.ones_like(df_corr,dtype=bool))
heatmap = plt.figure(figsize=(10,5))
axes = heatmap.add_axes([0.01,0.01,1,1])
sns.heatmap(df_corr,annot=True,ax=axes,cmap = "coolwarm",fmt='.2f',mask=mask)
plt.show()
```

```python
percentile33 = df['quality'].quantile(0.33)
percentile66 = df['quality'].quantile(0.66)
def kategory(a):
    if a>percentile66:
        return "High"
    elif a > percentile33:
        return "Medium"
    else:
        return "Low"
```

```python
df['quality cat'] = df['quality'].apply(kategory)
```

```python
df['quality cat'].unique()
```

```python
df.shape
```

```python
sns.scatterplot(data=df,x="density",y="fixed acidity",hue="quality cat")
plt.title("Fixed acidity vs Density over Wine quality")
plt.show()
```

```python
counts = df['quality cat'].value_counts()
plt.pie(df['quality cat'].value_counts(),labels=counts.index,autopct='%1.1f%%',startangle=90)
plt.title("Wine counts by category")
plt.show()
```
